```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/logistics_shipments_dataset - logistics_shipments_dataset.csv')
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
##
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean')
df['Cost'] = si.fit_transform(df[['Cost']])
```

```python
x1 = df.drop(columns=['Shipment_ID', 'Shipment_Date', 'Delivery_Date','Status'])
y = df['Status']
```

```python
##
x1 = pd.get_dummies(x1,columns=['Origin_Warehouse','Destination','Carrier'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
##
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
df['Status'].value_counts()
```

```python
## Fine tuning
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet'], # Changed 'none' to None
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [50,100,200,500],
    'l1_ratio': [0.5]
}
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
cl = GridSearchCV(
    LogisticRegression(),
    param_grid=param_grid,
    cv=6,
    scoring = 'accuracy',
    n_jobs=-1,
    verbose=2
)

cl.fit(x_test,y_test)
grid_y_pred = cl.predict(x_test)
```

```python
accuracy_score(y_test,grid_y_pred)
```

```python
cl.best_params_
```

```python
## F1-Score
from sklearn.metrics import f1_score
f1_score(y_test,grid_y_pred, average='weighted')
```
